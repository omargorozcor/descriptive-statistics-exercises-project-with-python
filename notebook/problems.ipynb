### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import matplotlib.pyplot as plt

###Distribución Normal
mu, sigma = 0, 1
normal_distribution = np.random.normal(size=100)
count, bins, ignored = plt.hist(normal_distribution, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='r')
plt.show()
````

````python
###Distribución Chi Cuadrado con 3 grados de libertad
chi_square = np.random.chisquare(3, size=100)

count, bins, ignored = plt.hist(chi_square, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='r')
plt.show()
````

````python
import random
from scipy import stats
from scipy.stats import iqr

##Distribuciones
distribucion_normal = np.random.normal(size=100)
chi_square = np.random.chisquare(3, size=100)

##distribucion_normal
##plt.hist(distribucion_normal)
##plt.show()
##res = stats.shapiro(distribucion_normal)
##res.statistic
##plt.hist(chi_square)
##plt.show()

# Metricas Normales (Solo son aplicables a la Distribución Normal)
media = np.mean(distribucion_normal)
desviacion = np.std(distribucion_normal)
print("Distribucion Normal")
print(f"La media es {media}")
print(f"La desviacion estandar es {desviacion}")


# Metricas No Normales (Solo son aplicables a las Distribuciones No normales)
mediana = np.median(chi_square)
rango_intercuartilico = iqr(chi_square)
print("Distribucion No Normal")
print (f"La mediana es {mediana}")
print (f"El rango intercuartilico es {rango_intercuartilico}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
###
data = [4,2,5,8,6]


def std(data):
    media = sum(data) / len(data)
    tamaño = len(data)
    sumatoria = 0
    for n in (data):
        valor1 = (n - media)**2 
        sumatoria += valor1
    desviacion_estandar = ((sumatoria)/tamaño)**0.5
    return desviacion_estandar


### Desviación estándar
print(f"La desviación estándar de los datos {data} es de {std(data)}")
print(f"Se puede comprobar que es correcto ya que la función .std del módulo numpy indica que es de {np.std(data)} ")

````
